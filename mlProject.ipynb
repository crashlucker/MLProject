<h3>Create a dataset with only predictors</h3>

```python
X= df.drop(columns=['date', 'Appliances', 'rv1', 'rv2'])
```

```python
X.head()
```

<h3>Looking for outliers</h3>
<p>zcore and IQR methods</p>

```python
from scipy.stats import zscore

z_scores = np.abs(zscore(y))

# if z-score > 3 => outlier
outliers = z_scores[z_scores > 3]
outliers.shape
```

```python
Q1 = y.quantile(0.25)
Q3 = y.quantile(0.75)
IQR = Q3 - Q1

outliers_iqr = df[(y < Q1 - 1.5*IQR) | (y > Q3 + 1.5*IQR)]
print(outliers_iqr.shape)
```

<h3>Shrink the target variable</h3>
<p>Need to shrink it to make it more smooth in presence of extreme values</p>

```python
y_capped = np.log(y)
```

```python
y_capped
```

```python
plt.hist(y_capped, bins=30)
plt.grid(True)
plt.show()
```

<h3>Split values in train and test</h3>
<p>Train and test have a proportion of 80:20 size</p>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_capped, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

<h2>Linear regression model</h2>
<p>Using pipeline for scaling features</p>

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

lin_model = Pipeline([
    ('scaler', StandardScaler()),
    ('linear', LinearRegression())
])

lin_model.fit(X_train, y_train)
```

<h3>Calculation of how good model predict</h3>

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

y_pred_lr = lin_model.predict(X_test)

mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = root_mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression")
print(f"MAE = {mae_lr}")
print(f"MSE = {mse_lr}")
print(f"RMSE = {rmse_lr}")
print(f"R2 = {r2_lr}")
```

<h2>Polynomial regression model</h2>

```python
from sklearn.preprocessing import PolynomialFeatures

poly_model = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures(degree=3, include_bias=False)),
    ('linear', LinearRegression())
])
```

<h3>Searching for the best parameters for this regression</h3>

```python
from sklearn.model_selection import GridSearchCV
param_grid_poly = {
    'poly__degree': [2, 3]       # polynomial degree
}

grid_poly = GridSearchCV(poly_model, param_grid_poly, cv=5, scoring='neg_mean_squared_error')
grid_poly.fit(X_train, y_train)
```

```python
grid_poly.best_params_
```

```python
y_pred_poly = grid_poly.predict(X_test)

mae_poly = mean_absolute_error(y_test, y_pred_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)
rmse_poly = root_mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)

print("Polynomial Regression")
print(f"MAE = {mae_poly}")
print(f"MSE = {mse_poly}")
print(f"RMSE = {rmse_poly}")
print(f"R2 = {r2_poly}")
```

<h2>Support Vector Regression</h2>

```python
from sklearn.svm import SVR

svr_model = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR(kernel='rbf', C=10, gamma='scale'))
])
```

```python
param_grid_svr = {
    'svr__C': [0.1, 10],
    'svr__epsilon': [0.01, 0.1]
}

grid_svr = GridSearchCV(svr_model, param_grid_svr, cv=5, scoring='neg_mean_squared_error')
grid_svr.fit(X_train, y_train)
```

```python
grid_svr.best_params_
```

```python
y_pred_svr = grid_svr.predict(X_test)

mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
rmse_svr = root_mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

print("Support Vector Regression")
print(f"MAE = {mae_svr}")
print(f"MSE = {mse_svr}")
print(f"RMSE = {rmse_svr}")
print(f"R2 = {r2_svr}")
```

<h3>Gradient Boost Regression</h3>

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(random_state=42)
```

```python
param_grid_gb = {
    'n_estimators': [50, 100],
    'learning_rate': [0.1, 0.2],
    'max_depth': [3, 5]
}

grid_gb = GridSearchCV(gb_model, param_grid_gb, cv=5, scoring='neg_mean_squared_error')
grid_gb.fit(X_train, y_train)
```

```python
grid_gb.best_params_
```

```python
y_pred_gb = grid_gb.predict(X_test)

mae_gb = mean_absolute_error(y_test, y_pred_gb)
mse_gb = mean_squared_error(y_test, y_pred_gb)
rmse_gb = root_mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print("Gradient Boosting")
print(f"MAE = {mae_gb}")
print(f"MSE = {mse_gb}")
print(f"RMSE = {rmse_gb}")
print(f"R2 = {r2_gb}")
```

<h2>Building model comparison</h2>
<p>Linear model made the worst prediction because we do not have linear relation. Strong underfit.</p>
<p>Polynomial Regression made an overfit</p>
<p>Everything else are goodfit or slightly overfit</p>

```python
models_comparison = []
models = {
    "Linear Regression": lin_model,
    "Polynomial Regression": grid_poly,
    "Support Vector Regression": grid_svr,
    "Gradient Boosting": grid_gb
}

for name, model in models.items():
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    models_comparison.append({
        'Model': name,
        'Train MAE': round(mean_squared_error(y_train, y_train_pred), 2),
        'Test MAE': round(mean_squared_error(y_test, y_test_pred), 2),
        'Train RMSE': round(root_mean_squared_error(y_train, y_train_pred), 2),
        'Test RMSE': round(root_mean_squared_error(y_test, y_test_pred), 2),
        'Train R2': round(r2_score(y_train, y_train_pred), 2),
        'Test R2': round(r2_score(y_test, y_test_pred), 2)
    })

df_models_comparison = pd.DataFrame(models_comparison)

print(df_models_comparison)
```

<h3>Searching for permutation importance in Gradient boost model</h3>

```python
from sklearn.inspection import permutation_importance

perm_gb = permutation_importance(grid_gb.best_estimator_, X_test, y_test, n_repeats=10, random_state=42, scoring='r2')

# Create dataframe
perm_importance = pd.Series(
    perm_gb.importances_mean,
    index=X.columns
).sort_values(ascending=False)

# Plot
plt.figure()
perm_importance.plot(kind='bar')
plt.title("Permutation Importance")
plt.ylabel("Decrease in R2")
plt.tight_layout()
plt.show()

# Print values
perm_importance
```
